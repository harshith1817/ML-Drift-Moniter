```python
# Converting SeniorCitizen column from numeric to categorical

churn_df["SeniorCitizen"]=churn_df["SeniorCitizen"].astype("object")
```

```python
numeric_cols=churn_df.select_dtypes(include=["int64", "float64"]).columns
categorical_cols=churn_df.select_dtypes(include=["object"]).columns
print("Number of numeric_cols:",len(numeric_cols))
print("Number of categorical_cols:",len(categorical_cols))
```

```python
binary_cols = [
    "Partner",
    "Dependents",
    "PhoneService",
    "PaperlessBilling"
]

for col in binary_cols:
    churn_df[col] = churn_df[col].replace({"Yes":1, "No":0})
churn_df["gender"] = churn_df["gender"].replace({"Female":1, "Male":0})
```

```python
for col in categorical_cols:
    print(f"{col}: {churn_df[col].unique()}")
```

```python
from sklearn.preprocessing import OneHotEncoder

multiple_cols=['InternetService', 
    'Contract', 
    'PaymentMethod',
    "OnlineSecurity",
    "OnlineBackup",
    "DeviceProtection",
    "TechSupport",
    "StreamingTV",
    "StreamingMovies",
    "MultipleLines"
]
churn_df = churn_df.reset_index(drop=True)
encoder= OneHotEncoder(sparse_output=False, handle_unknown='ignore')
for col in multiple_cols:
    encoder.fit(churn_df[[col]])
    encoded = encoder.transform(churn_df[[col]])
    encoded_df = pd.DataFrame(encoded, columns=encoder.get_feature_names_out([col]), index=churn_df.index)
    churn_df = pd.concat([churn_df, encoded_df], axis=1)
    churn_df = churn_df.drop(columns=[col])
```
